```python
def preproc(Djoined):
    Dmelt = pd.melt(Djoined, id_vars = filter(lambda x: re.match("^[^_]*$",str(x)),Djoined.columns.tolist()), 
                     value_vars =  filter(lambda x: re.match("^[A-Z]*_\d+$",str(x)),Djoined.columns.tolist()),
                     var_name = "measure",
                     value_name = "value")
    Dmelt["day"] = [int(x.split("_")[1]) for x in Dmelt["measure"]]
    Dmelt["measure"] = [x.split("_")[0] for x in Dmelt["measure"]]
    Dmelt = Dmelt[[not x for x in np.isnan(Dmelt.value)]]
    pivot = pd.pivot_table(Dmelt,["value"],["year","latitude","longitude","day"],["measure"])
    pivot = pivot.reset_index()
    pivot.columns = [[x for x in col if len(x)>0][-1] for col in pivot.columns.values]
    pivot.TM = pivot.TM/10
    return pivot
```
