```python
import sys
sys.path.append('../')
import dc_dashboard
from dc_dashboard import *
import dc_dashboard.HTTP as HTTP
```

```python
be = HTTP.HTTP_Backend("http://ion-21-14.sdsc.edu:8081/rdd", "tweets", 10000)
```

```python
dashboard = Dashboard(be, 
    [
        Dimension("date").
            add_group("hour",Group("function(d){return Math.floor(d/3600000)*3600000;}").reduce_count()).
            add_group("day",Group("function(d){return Math.floor(d/86400000)*86400000;}").reduce_count()),
        Dimension("userid"),
        Dimension("username"),
        Dimension("topics").add_group("count",Group().reduce_count()),
        Dimension("retweets").add_group("rounded",Group("function(d){return Math.ceil( (d > 0) ? Math.log10(d) : 0);}").reduce_count())
    ],
    [
        Layer(200),
        Plot("pie").data_source("retweets","rounded").title("Retweets Log 10").width(200),
        Plot("line").data_source("date","hour").title("Time Volume").width(800).config("domain",[1325376000000-3600000,1325376000000+86400000+3600000]),
        Layer(200),
        Plot("row").data_source("topics").title("Topics").width(1000),
        Layer(200),
        Plot("table").data_source("date").title("Example Tweets").width(1000).config("columns",["orig_date","username","retweets","topics","text"])
    ]
)
dashboard.show()
```
